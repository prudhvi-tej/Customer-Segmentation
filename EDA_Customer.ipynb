```python
from datetime import datetime
raw_df=data.copy()
data['Age']=2025-data['Year_Birth']
data['Education']=data['Education'].replace({'Basic':0,'2n Cycle':1,'Master':2,'PhD':3,'Graduation':4}).infer_objects().astype(int)
data['Marital_Status']=data['Marital_Status'].replace({'Married':1,'Together':1,'Single':0,'Divorced':0,'Widow':0,'Alone':0,'Absurd':0,'YOLO':0}).infer_objects().astype(int)
data['Children']=data['Kidhome']+data['Teenhome']
data['Family_size']=data['Marital_Status']+data['Children']+1
data['Total_Spending']=data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
data['Total_Promo']=data["AcceptedCmp1"]+ data["AcceptedCmp2"]+ data["AcceptedCmp3"]+ data["AcceptedCmp4"]+ data["AcceptedCmp5"]
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'],format="%d-%m-%Y")
today=datetime.today()
data['Days_as_Customer']=(today-data['Dt_Customer']).dt.days
data['Offers_Responded_To']=data['AcceptedCmp1']+data['AcceptedCmp2']+data['AcceptedCmp3']+data['AcceptedCmp4']+data['AcceptedCmp5']+data['Response']
data["Parental Status"] = np.where(data["Children"] > 0, 1, 0)
```

```python
data=data.rename(columns={"Marital_Status": "Marital Status","MntWines": "Wines","MntFruits":"Fruits",
                   "MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets",
                   "MntGoldProds":"Gold","NumWebPurchases": "Web","NumCatalogPurchases":"Catalog",
                   "NumStorePurchases":"Store","NumDealsPurchases":"Discount Purchases"})
data = data[["Age","Education","Marital Status","Parental Status","Children","Income","Total_Spending","Days_as_Customer","Recency","Wines","Fruits","Meat","Fish","Sweets","Gold","Web","Catalog","NumWebVisitsMonth"]]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def compute_vif(conti_features,data):
    X=data[conti_features].copy()
    X['intercept']=1
    vif=pd.DataFrame()
    vif['Variable']=X.columns
    vif['VIF_val']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    vif=vif[vif['Variable']!='intercept']
    return vif

conti_features=con_feat.copy()
conti_features.remove('Total_Spending')
compute_vif(conti_features,data)
```
